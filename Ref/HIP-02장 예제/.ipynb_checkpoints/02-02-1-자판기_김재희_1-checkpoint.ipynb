```python
# 김재희씨 작업

import sys

class vend_machine:
    def __init__(self):
        self.product = {'커피': 200, '라떼': 300, '생수': 400}
        self.in_money = 0
        self.choice = 0
        self.cnt = 0

    def money(self):
        self.in_money = int(input("돈을 넣어주세요 : "))

    def add_money(self):
        while True:
            check = input("돈을 추가하시겠습니까?(Y/N) : ")
            if check == 'Y':
                self.in_money += int(input("추가금액 : "))
                break
            elif check == 'N':
                self.get_describe()
                break
            else:
                print("잘못된 입력입니다.")

    def select(self):
        print("%d원이 있습니다." % self.in_money)
        choice = int(input("""원하는 음료를 고르시오.\n1.커피 2.라떼 3.생수   0.종료\n 200원 300원 400원\n: """))
        if 1 <= choice <= len(self.product):
            self.choice = list(self.product.keys())[choice - 1]
            while True:
                self.cnt = int(input("수량 : "))
                can_buy = (self.in_money // self.product[self.choice])
                if self.cnt > can_buy:
                    print("최대 가능 수량은", can_buy, "개입니다.")
                elif can_buy == 0:
                    self.add_money()
                else:
                    break
        elif choice == 0:
            self.get_describe()
        else:
            print("다시 선택해주세요")
            self.select()

        self.get_drink()

    def get_drink(self):
        self.in_money -= self.product[self.choice] * self.cnt
        print(self.choice, self.cnt, "잔을 드립니다.\n남은돈 : ", self.in_money)

        choice = int(input("1. 계속구매  2.종료\n: "))
        if choice == 2:
            sys.exit()


    def get_describe(self):
        print('잔돈이 %d원 반환되었습니다.' % self.in_money)
        sys.exit()

vm = vend_machine()
vm.money()
while True:
    if vm.in_money < min(vm.product.values()):
        print("구매하기엔 돈이 부족합니다.")
        vm.add_money()

    vm.select()
```

```python
# 최기석씨 소스 
class vending_machine:

    def __init__(self):
        self.product = (['생수', 300], ['코코팜', 1000], ['모구모구', 1300])
        self.money = 0

    def insert_money(self):
        print()
        self.money += int(input("돈을 넣어주세요: "))
        print("잔액: ", self.money)

    def choice_product(self):
        print()
        cnt = 0
        for i in self.product:
            cnt += 1
            print(cnt, ": ", i)
        product_name = input("상품 번호를 선택해주세요(돈을 더 넣으시려면(0)): ")
        product_name = int(product_name)
        if str(type(product_name)) != "<class 'int'>":
            print("잘못선택하셨습니다.")
            return 1
        elif product_name > len(self.product):
            print(len(self.product), "번호 내에서 선택해 주세요")
            return 1
        else:
            if product_name == 0:
                return 0
            else:
                print(self.product[product_name - 1][0], "상품이 선택되었습니다.")
                self.money = self.money - self.product[product_name - 1][1]
                if self.money > 0:
                    print("잔액이 ", self.money, " 남았습니다.")
                    return 1
                else:
                    self.money + self.product[product_name - 1][1]
                    print("잔액이 모자랍니다.")
                    return 1


vm = vending_machine()
step = 0
while True:
    if step == 0:
        vm.insert_money()
        step = 1
    elif step == 1:
        step = vm.choice_product()
    elif step == 2 
        print("작동을 멈춥니다.")
        break 
```
