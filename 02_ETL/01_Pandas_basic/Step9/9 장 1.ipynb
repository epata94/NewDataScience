##  예제 9-1 그래프 틀과 세부 그래프 표시하기

```python
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 30)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
fig, axes = plt.subplots(2,2,figsize=(4,4))
axes[0,0].plot(x,y, 'r')
axes[0,1].plot(x,y, 'g')
axes[1,0].plot(x,y, 'b')
axes[1,1].plot(x,y, 'k')
for ax in axes.ravel():
    ax.margins(0.05)
    
fig.suptitle(" four graph ")

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(x,y, '--')
ax2.plot(x,y, 'o')
ax3.plot(x,y, "^")
ax4.plot(x,y, '*')

ax1.set_title(" hline ")
ax2.set_title(" circle ")
ax3.set_title(" triangle_up ")
ax4.set_title(" star ")

plt.show()
```

```python
import tensorflow as tf
```

```python
a= np.array([1,2,3,4,5,6], dtype=np.float32)
b= np.array([2,3,4,4,7,7], dtype=np.float32)
c=tf.matmul(a.reshape(3,2),b.reshape(2,3))
```

```python
sess=tf.Session()
a=sess.run(c)
for i, row in enumerate(a):
    x =np.arange(len(row))
    plt.bar(x+(0.25*i),row,color= (lambda i: ("b g r".split())[i])(i), width=0.25)
    
plt.show()
print(a)
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```
