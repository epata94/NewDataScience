## 그래프 영역 채우기

그래프의 특정 영역을 색상으로 채워서 강조
* fill_between() - 두 수평 방향의 곡선 사이를 채웁니다.
* fill_betweenx() - 두 수직 방향의 곡선 사이를 채웁니다.
* fill() - 다각형 영역을 채웁니다.

### 기본 사용 - fill_between()

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
```

### 기본 사용 - fill_betweenx()

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)

# x축 영역을 조정하여 공백 제거
plt.xlim([1, 4])

plt.show()
```

### 다각형 영역 채우기 - fill()

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='lightgray', alpha=0.5)

plt.show()
```

### 타이틀 정보 얻기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
title_right = plt.title('Left Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
title_left = plt.title('Right Graph Title', fontdict=title_font, loc='left', pad=20)

print(f'왼쪽 타이틀 위치: {title_left.get_position()}')
print(f'왼쪽 타이틀 이름: {title_left.get_text()}\n')

print(f'오른쪽 타이틀 위치: {title_right.get_position()}')
print(f'오른쪽 타이틀 위치: {title_right.get_text()}')

plt.show()
```

## 수평선/수직선 표시하기

### 수평선 그리기 - axhline([y위치, x시작점(퍼센트), x종료점(퍼센트)], 기타 스타일 옵션)

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)

plt.show()
```

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.5, 1, color='lightgray', linestyle='--', linewidth=2)

plt.show()
```

### 수평선 그리기 - hlines([y위치, x시작점, x종료점], 기타 스타일 옵션)

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
# x 스케일에 맞추어 수치 입력
plt.hlines(-0.62, 0.5, 3.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```
