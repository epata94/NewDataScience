## 스타일 지정

```python
                                     #  <= Format String
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
# axis 함수로 축의 범위 지정 [xmin, xmax, ymin, ymax]
plt.axis([0, 6, 0, 20])
plt.show()
```

## 여러 개의 그래프 그리기

```python
import numpy as np
t = np.arange(0., 5., 0.2)
# 0에서 5사이 0.2씩 증가
t
```

```python
# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

## 변수 활용

```python
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

## 축 설정
### 레이블 사용

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
# X축 레벨
plt.xlabel('X-Label')
# Y축 레벨
plt.ylabel('Y-Label')
plt.show()
```

### 여백 설정

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
# labelpad 인자로 여백 지정 (단위 pt)
plt.xlabel('X-Axis', labelpad=15)
plt.ylabel('Y-Axis', labelpad=20)
plt.show()
```

### 폰트 설정

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
plt.show()
```

### 위치 설정 

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
# x축 레이블 위치 지정 (‘left’, ‘center’, ‘right’)
plt.xlabel('X-Axis', loc='right')
# y축 레이블 위치 지정 (‘bottom’, ‘center’, ‘top’)
plt.ylabel('Y-Axis', loc='top')
plt.show()
# 변경해 보세요.
```

## 범례 표시하기

```python
# Step1: label에 범례명을 입력한다.
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# Step2: legend함수를 호출한다.
plt.legend()

plt.show()
```

### 범례 위치 지정

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# 좌측하단
plt.legend(loc=(0.0, 0.0))
# 중간
# plt.legend(loc=(0.5, 0.5))
# 우측산단
# plt.legend(loc=(1.0, 1.0))

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='upper left')

plt.show()
```

## 열개수 지정

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# loc='best' 최적의 공간을 찾아준다. 디폴트 옵션은 1열로 범례를 표시한다.
plt.legend(loc='best')          # ncol = 1
#2열로 범례를 표시한다.
# plt.legend(loc='best', ncol=2)    # ncol = 2

plt.show()
```

### 폰트 크기 지정하기 

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=14)

plt.show()
```

### 범례 테두리

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
# frameon: 테두리 생성여부
# shadow: 텍스트 상자에 그림자 여부
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)

plt.show()
```

### 축 범위

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5])      # X축의 범위: [xmin, xmax]
plt.ylim([0, 12])     # Y축의 범위: [ymin, ymax]

plt.show()
```
