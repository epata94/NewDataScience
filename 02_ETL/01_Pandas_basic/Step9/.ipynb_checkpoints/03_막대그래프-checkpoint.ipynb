## 기본 사용

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)

plt.show()
```

### 색상 지정하기

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, color='y')
# plt.bar(x, values, color='dodgerblue')
# plt.bar(x, values, color='C2')
# plt.bar(x, values, color='#e35f62')
plt.xticks(x, years)

plt.show()
```

### 바마다 개별 색 지정

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

# 색상이름을 리스트로 지정하면 순서대로 바에 할당
plt.bar(x, values, color=colors)
plt.xticks(x, years)

plt.show()
```
