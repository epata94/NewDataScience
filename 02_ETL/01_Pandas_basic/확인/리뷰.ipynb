```python
import pandas as pd
import numpy as np

musicData = pd.read_csv('./data/musicData.csv', encoding='UTF-8')
musicData2 = pd.read_csv('./data/musicData2.csv', encoding='UTF-8')
```

```python
music_i = musicData.set_index(['Date'])
music_i
```

```python
music2_i = musicData2.set_index('Date')
music2_i
```

<hr> 
<hr>

# 문제 1
- 상단의 데이터프레임 2개를 inner join 하시오.
- 중복 되는 열 이름은 구별 되게 출력 하시오

```python
music_join = music_i.join(music2_i,how='inner',rsuffix='_new')
music_join.head()
```

 <hr> 
<hr>

# 문제 2
- 문제 1의 결과를 이용하여 Album Title 과 Artist Name, Rating을 고정열로 설정하여 Genre 열 두개를 병합 처리하여 출력 하시오.
- 변수명은 자유롭게 설정 하되, melt 함수 사용 하시오

```python
music_genre = music_join.melt(id_vars = ['Album Title','Artist Name','Rating'], var_name = "melt_Genre", value_name = "genre", value_vars=['Genre','Genre_new'])

music_genre
```

# 보너스 문제
- 문제 2의 결과에 Rating 열을 기반으로 순위 열을 만드시오.
- 단, 동점인 데이터는 데이터 상에서 먼저 나타나는 관측치부터 우선 순위를 부여 하고 순위로 오름차순으로 정렬하여 출력 하시오.

```python
music_genre['순위'] = music_genre['Rating'].rank(method='first',ascending=False).astype(int)
music_genre.sort_values(by='순위')
```
