# 뮤직데이터를 장르와 년도로 멀티인덱싱하고  랭킹별, 점수별 평균과 최솟값, 최대값을 구하시오

```python
group_cols = ['Genre','Year']
```

```python
agg_cols=['Rating','Music Count']
```

```python
agg_music = music.groupby(group_cols)[agg_cols].agg(['mean','max','min'])
```

```python
agg_music
```

# 피봇을 이용해 인덱스를 아티스트 이름으로, 컬럼을 년도로, 값은 평점의 평균으로 재구조화 하시오 

```python
music_pivoted=music.pivot_table(index='Artist Name', columns='Year', values='Rating', aggfunc='mean' )
music_pivoted
```

# 널값이 있는 행을 삭제하고 2017년과 2018년에 모두 노래를 낸 가수의 수를 구하시오

```python
both_music=music_pivoted.dropna()
```

```python
both_music
```

```python
len(both_music)
```

# 앞서 구한 테이블에 2017년의 성적과 2018년의 성적의 평균을 구하는 열을 구하시오

```python
both_music['평균'] = both_music.eval(both_music[2017] + both_music[2018])/2
```

```python
both_music
```

```python
both_music.sort_values(by="평균")
```

```python
both_music.to_csv("example_both_music.csv",encoding='cp949')
```
