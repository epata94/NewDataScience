```python
for i in market_df.index:
    market_df.iloc[i,12] = market_df['소재지지번주소'].str.split()[i][0]
```

- Split 함수로 소재지지번주소 열의 값을 공백을 기준으로 잘라 자른 값의 0번째인 값을 '시도'열에 삽입

```python
market_df['시도']
```

```python
market_df['시도'].value_counts() # 시각화
```

```python
market_df['시도'].replace('화성시','경기도',inplace=True)
market_df['시도'].value_counts()
```

```python
city_count = market_df['시도'].value_counts()
```

```python
city_count = pd.DataFrame(city_count)
city_count.reset_index(inplace=True)
city_count.columns = ['시도','시장갯수']
```

```python
city_count
```

```python
 city_count2 = city_count.sample(frc=1).reset_index(drop=True)
```

```python
plt.figure(figsize=(20,6))
explode = (0, 0, 0, 0, 0, 0,0.2,0,0.2,0,0,0,0,0,0.3,0,0)
wedgeprops={'width': 0.8, 'edgecolor': 'w', 'linewidth': 2}
color=sns.color_palette('summer',len(city_count.index))
plt.subplot(1,2,1)
plt.pie(city_count2['시장갯수'], labels=city_count2['시도'],autopct='%.1f%%',explode=explode,wedgeprops=wedgeprops, shadow=True)
plt.title('전국 시장 분포도')
plt.subplot(1,2,2)
plt.ylim(100,250)
sns.barplot(data=city_count.head(),x="시도",y="시장갯수")
plt.title('전국 시장 갯수 순위')
plt.show()
```

* 처음 예상에는 마트가 많이 들어와있는 서울이 전통시장이 가장 적을꺼라고 생각했지만
* 데이터 분석 결과 서울특별시에 인구 밀도수가 높아서 그에 비례하여 시장 수 또한 높았다.

* 전체 상품권 종류 중 전통 시장에서 사용 가능한 상품권은 압도적으로 온누리 상품권이 높다.

```python
year_count = market_df['개설연도'].value_counts()
year_count = pd.DataFrame(year_count)
year_count.reset_index(inplace=True)
year_count.columns=['년도','증가갯수']
year_count
```

```python
index=0

for i in year_count['년도']:
    if len(i) != 4:
        print(i)
        year_count.iloc[index,0] = year_count.iloc[index,0].split('-')[0]
    index += 1
```

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=70)

sns.lineplot(x='년도',y='증가갯수',data=year_count.sort_values(by='년도',ascending=True))
```
