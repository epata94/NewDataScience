### 각 열의 의미
* Rank : 글로벌 매출액 기준 게임 순위
* Game Title : 게임 제목
* Platform : 게임이 발매된 플랫폼
* Year : 게임이 발매된 해
* Genre :  게임의 장르
* Publisher :  게임의 퍼블리셔
* North America : 북미에서의 게임 판매량
* Europe : 유럽에서의 게임 판매
* Japan : 일본에서의 게임 판매량
* Rest of World : 다른 나라에서의 게임 판매량
* Global : 게임의 글로벌 총 매출액입니다
* Review : 게임의 리뷰 점수입니다
