### 1-1. restaurant에서 판매하는 아이템 중 어떤 아이템이 가장 많이 팔리는가?  

```python
restaurant_item = restaurant['item_name'].value_counts()[:10]
restaurant_item
```

##### => 가장 많이 팔리는 아이템으로 부터 10가지를 slicing 하여  Top10 을 가져왔다

```python
restaurant_item_index = restaurant_item.index
restaurant_item_index
```

```python
plt.pie(restaurant['item_name'].value_counts()[:10],labels=restaurant_item_index, autopct='%.1f%%')
plt.show()
```

##### => 판매 비율을 시각화 하기 위하여 pie chart 로 나타냈다

### 1-2. 가장 많이 판매되는 아이템이 가장 높은 수익율을 나타내는가? 

```python
restaurant_item_sales = restaurant.groupby(restaurant['item_name'])['item_price'].sum().to_frame()
top = restaurant_item_sales.sort_values('item_price', ascending=False)[:10]
top
```

##### => 가장 높은 수익을 내는 아이템을 나타냈다

```python
idx = top.index
colors = sns.color_palette('Set3',len(idx))
top.plot.bar(y='item_price', color=colors)
plt.xticks(rotation = 90)
```

##### => 메뉴별 매출액이 높은 순서대로 시각화 하였다

Step1의 결과

## 3-1. 전체 테이블에서 'Canned'를 포함한 아이템을 검색

```python
restaurant[restaurant['item_name'].str.contains('Canned')]
```

=> 레스토랑의 'Canned'를 contains 함수로 검색해 보니 'item_name'에 'Canned Soda'와 'Canned Soft Drink'를 발견  
=> 'Canned Soda'와 'Canned Soft Drink' 는 같은 아이템이라고 추정

## 3-2. 아이템 리스트 추출

```python
Canned_columns = restaurant[['quantity','item_name','choice_description','item_price']]
Canned_columns
```

```python
Canned= Canned_columns[restaurant['item_name'].str.contains('Canned')]
Canned
```

```python
Canned_list=Canned['choice_description'].drop_duplicates().tolist()
Canned_list
```

=> 'Canned'를 contains 함수를 활용하여 1차 분리 하였고, 'choice_description'의 중복값을 제거하여  
'Canned_* '의 메뉴를 가져왔다  
=> Cola 의 경우도 Coke 와 Coca Cola로 나뉜걸 볼 수 있다  
=> 이는 pepsi와 Coca Cola 의 차이라고 추정된다 추후 확인이 필요하다.

## 3-5. 1.09로 판매되었던 Canned Drink를 모두 1.25로 팔았을 때 생기는 수익률

```python
Canned_total_price = Canned['item_price'].sum()
Canned_total_price
```

```python
low_count = len(Canned[Canned['item_price'] == 1.09])
low_count
```

```python
Canned_profit = (1.25*low_count)-(1.09*low_count)
Canned_profit
```

```python
expected_return = (Canned_profit/Canned_total_price)*100
expected_return
```

=> Canned Drink의 수익의 약 2.33% 정도 증대를 기대할 수 있다

# Step3 결과
* 레스토랑의 'Canned'를 contains 함수로 검색해 보니 'item_name'에 'Canned Soda'와 'Canned Soft Drink'를 발견  
* 추후 매장의 확인이 필요해 보이나 같은 메뉴가 2개로 추정된다  
* 어떠한 이유로 인지 'Canned Soda'의 가격은 개당 1.09 이고, 'Canned Soft Drink'의 가격은 1.25 이다  
* 이는 고객에게 메뉴의 혼동을 유발할 수 있고, 컴플레인을 발생할 수 있는 데이터로 보인다  
* 1.25와 1.09는 약 7.8%의 차이를 보여준다  
* 이를 세트메뉴의 할인율로 잡으려 한다
