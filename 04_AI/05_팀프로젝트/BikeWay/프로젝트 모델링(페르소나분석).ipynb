# 프로젝트 모델링

## SNS 키워드 단어 빈도 기반 사용자 페르소나 분석

### 페르소나 정의

* 사용자 모델링에서 페르소나는 페르소나 기법을 의미한다.
* 어떤 제품 혹은 서비스르 사용할 만한 목표 인구 집단 안에 있는 다양한 사용자 유형들을 대표하는 가상 인물을 구체화하는 것을 말한다.
* 가상의 인물을 묘사하고 그 인물의 배경과 환경 등을 설명하는 문서로 꾸며진다.
* 이 때 가상의 이름, 목표, 평소에 느끼는 불편, 그 인물이 가지는 니즈 등으로 구성된다.

### 데이터 셋팅

#### SNS 키워드를 바탕으로 라이딩 유저 데이터 수집

* 네이버 뉴스, 네이버 블로그에서 '자전거','라이딩'을 포함하는 뉴스와 게시물의 본문을 크롤링
* 네이버 뉴스 : 1페이지 ~ 50페이지에 해당되는 데이터 수집
* 네이버 블로그 : 약 500개의 블로그 해당되는 데이터 수집

#### 데이터 빈도를 바탕으로 텍스트 시각화

#### 데이터 기반의 분석결과 

* 데이터 기반의 분석결과 코스에 관한 관심도가 가장 높았고, 날씨와, 제품, 사진등의 키워드 들을 확인할 수 있었다.

### 페르소나 셋팅

#### 1. 오운완

* 프로파일명 : 오운완
* 프로파일정보  
    * 나이 : 20대 중후반
    * 직업 : 회사 초년생
    * SNS활용도 : 매일
    * 자전거 관심도 : 중 
    * 특이사항 : 최근 헬스장에 지루함을 느껴 다른 운동을 찾는 중 

* 시나리오 
    * 평소 워라벨을 중요시하고, 쉬는날 or 퇴근 후 헬스장에서 운동을 하면서 SNS에 인증샷을 올림
    * 헬스장을 다닌지 약 6개월이 넘어가 지루해하고 있어 다른 운동을 알아보는중 
    * SNS에 예쁜 사진을 올리는것이 삶의 낙이자 취미생활
    * SNS를 위해서라면 돈쓰는것을 아끼지 않음

#### 2. 주말엔 라이딩

* 프로파일명 : 주말엔 라이딩
* 프로파일정보  
    * 나이 : 40대
    * 직업 : 회사 부장급
    * SNS활용도 : 한달에 한번
    * 자전거 관심도 : 상 
    * 특이사항 : 자전거와 함께하는 주말이 기다려짐 

* 시나리오 
    * 주말에는 거의 대부분 자전거 타는 것에 매진
    * 자전거를 타고 항상 맛있는 집에서 밥을 먹는것에 삶의 행복을 느낌
    * 날씨를 확인하지 않고 라이딩을 하다가 비를 맞아 고생한 기억이 있음
    * 한번은 자전거 부품이 고장나 고생했던 기억도 있음
    * 자전거 라이딩은 좋아하나 다양한 정보를 얻는것에 피로감을 느낌

#### 3. 확찐자

* 프로파일명 : 확찐자
* 프로파일정보  
    * 나이 : 20~30대
    * 직업 : 대리급 직원
    * SNS활용도 : 주 3~4회
    * 자전거 관심도 : 하
    * 특이사항 : 코로나 19 이후 집에 머무르는 시간이 많아 살이 부쩍 오름 

* 시나리오 
    * 코로나 19 이후 집 회사를 반복해 보통 체형에서 살이 급격히 찜
    * 운동에는 1도 관심이 없고 시간도 투자하기 싫어함
    * 보유 차량이 없어 이동수단은 대중교통을 이용, 서울 거주

#### 4. This is Korea

* 프로파일명 : This is korea
* 프로파일정보  
    * 나이 : 10 ~ 20대
    * 직업 : 프리랜서
    * SNS활용도 : 주 5회
    * 자전거 관심도 : 하
    * 특이사항 : 외국인, 한국에 여행온 한국의 문화와 다양한 정보를 얻고 싶어함 

* 시나리오 
    * BTS로 한국과 한국문화가 궁금해서 서울로 여행온 외국인
    * 한국의 다양한 장소와 음식을 궁금해함
    * 장소별 이동수단의 정보가 미흡함
    * 한국어를 공부해 한국인과의 원활한 커뮤니케이션이 가능

### 페르소나 분석

#### 1. 오운완(잠재유저)

* 프로파일명 오운완의 경우 평소 운동하는 것을 즐겨하며 다양한 운동에 관심을 가지고 있다.
* SNS의 사진을 업로드하며, 좋아요를 많이 받을때 행복함을 느낀다.

-> 신규 자전거 유저를 위해 자전거를 추천, 비교해주는 페이지로 유저의 유입을 기대  
   다양한 코스와 코스 근처의 맛집,플레이스 등을 추천해준다

#### 2. 주말엔 라이딩 (충성유저)

* 프로파일명 '주말엔 라이딩'의 경우 주말에는 빠짐없이 자전거 라이딩을 즐긴다.
* 자전거 라이딩은 좋아하나 라이딩에 필요한 다양한 정보들을 수집하는데 피로감을 느낀다.

-> 한페이지에 코스와 맛집 날씨를 보여줄 수 있는 All in One 페이지를 제작해 유저 확보한다.

#### 3. 확찐자 (관심이 없는 유저)

* 프로파일명 '확찐자'의 경우 평소 운동에 관심이 없었고, 코로나19 이후 더욱 활동이 적어져 최근 살이 찐 경우이다.
* 서울에 거주하며 이동수단은 대중교통을 이용한다.

-> 서울시 공유 자전거(따릉이)를 통해 출퇴근 길에 자전거를 탈수있도록 유입  
집 근처의 따릉이 대여소의 정보를 실시간으로 제공하여 유저의 유입을 기대  
더 나아가 따릉이 대여소의 시계열 예측을 통하여 특정시간,날 대여소가 붐비는지 데이터 제공 

#### 4. Is this Korea ?

* 프로파일명 'Is this Korea?'의 경우 BTS로 한국과 한국 문화가 궁금해 서울로 여행온 외국인이다.
* 서울로 여행계획을 준비중이나, 한국음식과 다양한 장소에 대해 정보가 없는 상황이다.

->  
    * 서울의 다양한 음식점 정보를 제공하여 유저의 유입을 기대한다.
    * 이동 장소가 짧을 경우 따릉이 공유 자전거의 사용을 유도해본다.
