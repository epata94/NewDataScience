### 시계열 데이터 시각화

```python
titles = [
    "Pressure",
    "Temperature",
    "Temperature in Kelvin",
    "Temperature (dew point)",
    "Relative Humidity",
    "Saturation vapor pressure",
    "Vapor pressure",
    "Vapor pressure deficit",
    "Specific humidity",
    "Water vapor concentration",
    "Airtight",
    "Wind speed",
    "Maximum wind speed",
    "Wind direction in degrees",
]

feature_keys = [
    "p (mbar)",
    "T (degC)",
    "Tpot (K)",
    "Tdew (degC)",
    "rh (%)",
    "VPmax (mbar)",
    "VPact (mbar)",
    "VPdef (mbar)",
    "sh (g/kg)",
    "H2OC (mmol/mol)",
    "rho (g/m**3)",
    "wv (m/s)",
    "max. wv (m/s)",
    "wd (deg)",
]

colors = [
    "blue",
    "orange",
    "green",
    "red",
    "purple",
    "brown",
    "pink",
    "gray",
    "olive",
    "cyan",
]

date_time_key = "Date Time"


def show_raw_visualization(data):
    time_data = data[date_time_key]
    fig, axes = plt.subplots(
        nrows=7, ncols=2, figsize=(15, 20), dpi=80, facecolor="w", edgecolor="k"
    )
    for i in range(len(feature_keys)):
        key = feature_keys[i]
        c = colors[i % (len(colors))]
        t_data = data[key]
        t_data.index = time_data
        t_data.head()
        ax = t_data.plot(
            ax=axes[i // 2, i % 2],
            color=c,
            title="{} - {}".format(titles[i], key),
            rot=25,
        )
        ax.legend([titles[i]])
    plt.tight_layout()


show_raw_visualization(df)
```

# 시계열 처리 모델

### prophet을 사용하기 위해서는 다음과 같이 열 이름을 변경 : 'ds', 'y'

```python
df_t = df2015_t.reset_index()
df_t.head()
```

```python
df_t.rename(columns={"Date_Time":"ds","온도":"y"},inplace='True')
df_t.head()
```

```python
from neuralprophet import NeuralProphet, set_log_level
```

```python
prophet_m= NeuralProphet(
    learning_rate=0.1, 
)
```

```python
metrics = prophet_m.fit(df_t,freq="H")
```

```python
train_pred = prophet_m.predict(df_t)
# 30*24*6 : 1시간에 6번 저장 기준 24시간 수집되어 최근 30일 적용
fig1 = prophet_m.plot(train_pred[-30*24*6:]) 
param = prophet_m.plot_parameters()
```

### 하이퍼파라메터 튜닝

```python
m2 = NeuralProphet(

growth='discontinuous', # 추세 유형 설정(linear, discontinuous, off 중 선택 가능)

yearly_seasonality=True, #년간 계절성 설정

weekly_seasonality=True, #주간 계절성 설정

daily_seasonality=True, #일간 계절성 설정

batch_size=64,#배치 사이즈 설정

epochs=100,#학습 횟수 설정

learning_rate=0.1, # 학습률 설정

n_lags= 1, #1시간 전 데이터로 설정

)
```

### Prophet 학습을 위한 추가 데이터 표준화

```python
import numpy as np
df2 = df_t.copy(deep=True)
df2["l"] = np.append(0, df2["y"].values[1:] - df2["y"].values[:-1])
df2=df2[['ds','y','l']]
df2.tail(3)
```

```python
m2 = m2.add_lagged_regressor(names=["l"], normalize="minmax") 
```

```python
metrics = m2.fit(df2,freq="H")
```

```python
train_pred = m2.predict(df2)
#fig = m.plot(train_pred)
fig1 = m2.plot(train_pred[-30*24*6:])
# fig2 = m.plot(train_pred[-7*24*6:])a
#comp = m.plot_components(train_pred)
param = m2.plot_parameters()
```
